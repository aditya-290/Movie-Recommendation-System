```python
import re
def remove_re(text):
    text = re.sub(r'&quot;', '', text)
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    
    return text

movie_data['title'] = movie_data['title'].apply(remove_re)
```

```python
movie_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tvf = TfidfVectorizer(min_df = 3,max_features=None,strip_accents = 'unicode',analyzer = 'word',token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')


movie_data['genres'] = movie_data['genres'].fillna('')
genre_str = movie_data['genres'].str.split(',').astype(str)

tvf_matrix = tvf.fit_transform(genre_str)
```

```python
tvf_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tvf_matrix, tvf_matrix)
```

```python
indices = pd.Series(movie_data.index, index=movie_data['title']).drop_duplicates()
print(indices)
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]
    
    #
#     pickle.dump(sig_scores, open('model/sig_scores.pkl','wb'))
    

    # Top 10 most similar movies
    return pd.DataFrame({'Movie name': movie_data['title'].iloc[movie_indices].values})
    
```
